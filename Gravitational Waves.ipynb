# Detecting Exo Planets.

```python
from lightkurve import search_targetpixelfile
tpf = search_targetpixelfile('KIC 6922244', author = "Kepler", cadence = "long", quarter = 4).download()
lc = tpf.to_lightcurve(aperture_mask = tpf.pipeline_mask)
```

```python
tpf.plot()  # This visuals is showing us the light from the star.
```

```python
tpf.flux.shape
```

```python
lc.meta["MISSION"]
```

```python
lc.astropy_time.utc.iso
```

```python
lc.plot()   # Visual is showing brightness of the star periodically.
```

```python
flat_lc = lc.flatten(window_length=401)
flat_lc.plot()
```

```python
periodogram = flat_lc.to_periodogram(minimum_period=1, maximum_period=4, oversample_factor=30)
periodogram.plot()
```

```python
best_fit=periodogram.period_at_max_power
print(best_fit)
```

```python
folded_lc = flat_lc.fold(period = 3.5225)
folded_lc.plot()
```
